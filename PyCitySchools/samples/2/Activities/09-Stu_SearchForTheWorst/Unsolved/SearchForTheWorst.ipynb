```python
# Import Dependencies
import pandas as pd
```

```python
# Create reference to CSV file
csv_path = "Resources/SFO_Airport_Utility_Consumption.csv"

# Import the CSV into a pandas DataFrame
consumption_df = pd.read_csv(csv_path)
consumption_df
```

```python
# Collect a list of all the unique values in "Utility"
list(consumption_df["Utility"].unique())
```

```python
consumption_df["Utility"].value_counts().to_frame()
```

```python
# Looking only at Electricity Consumption with "Tenant" owner


consumption_df.loc[(consumption_df["Utility"]=="Electricity") & (consumption_df["Owner"]=="Tenant")]
```

```python
consumption_df[(consumption_df["Utility"]=="Electricity") & (consumption_df["Owner"]=="Tenant")]
```

```python
consumption_df["Usage"] = consumption_df["Usage"]/10000
```

```python
# Sort the DataFrame by the values in the "Usage" column to find the worst day
df2 = consumption_df.sort_values("Usage", ascending=False)
# Reset the index so that the index is now based on the sorting locations
df2 = df2.reset_index(drop=True)
df2
```

```python
# Save all of the information collected on the worst day
df2.to_csv("Resources/SFO_Airport_Utility_Consumption_worst day.csv", index=False)
```
