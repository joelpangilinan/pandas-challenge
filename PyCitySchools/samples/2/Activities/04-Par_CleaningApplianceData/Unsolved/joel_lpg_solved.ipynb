```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located
lpg_file = "../Resources/dga_lpg.csv"

# Import the data into a Pandas DataFrame
lpg_df = pd.read_csv(lpg_file, encoding="utf-8")
lpg_df.head()
```

```python
# Reduce to columns that are in English
lpg_reduced_df = lpg_df[["Part","Type","Brand","Model","Other Information","Place of Manufacture","Applicant","Telephone Number","Approval Expiry Date"]]
lpg_reduced_df.head()
```

```python
# look for missing values
lpg_reduced_df.count()
```

```python
# drop null rows
lpg_drop_rows_df = lpg_reduced_df.dropna(how="any")
```

```python
# verify counts
lpg_drop_rows_df.count()
```

```python
# Check to see if there are any values with mispelled or similar values in "Applicant"
#list(lpg_drop_rows_df["Applicant"])
lpg_drop_rows_df["Applicant"].unique()
```

```python
# Combine similar applicants together
lpg_combined_applicants_df = lpg_drop_rows_df.replace(
    {"Crown Gas Stoves (Holdings) Company Limited": "Crown Gas Stoves Co., Ltd.", 
     "Sun Kee LP Gas Co.": "Sun Kee LP Gas Co. Limited"})
lpg_combined_applicants_df.head()
```

```python
# Check to see if you combined similar applicants correctly in "Applicant"
lpg_combined_applicants_df["Applicant"].unique()
```

```python
# Create a new DataFrame that looks into a specific Place of Manufacture
lpg_manufacturer_country_df = lpg_combined_applicants_df.loc[lpg_combined_applicants_df["Place of Manufacture"] == "France"]
lpg_manufacturer_country_df
```
